```python
import numpy as np
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
seed = 2020
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
data_set = np.loadtxt("../dataset/ThoraricSurgery.csv", delimiter=",")
data_set
```

```python
# 환자의 기록과 수술 결과를 X와 Y로 구분하여 저장
# 17개의 input  hiddenlayer(은닉층) 34개
x_data = data_set[:,0:17]
y_data = data_set[:, 17]
x_data.shape, y_data.shape
```

```python
# 딥러닝 모델 설정
model = Sequential()
model.add(Dense(24, input_shape = (17,), activation ='relu'))
model.add(Dense(8, input_shape = (17,), activation ='relu'))
model.add(Dense(1, activation = 'sigmoid'))

model.summary()
```

```python
# 딥러닝 환경 설정 (오차함수, 최적화함수)
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# 딥러닝 실행
model.fit(x_data, y_data, epochs=500, batch_size=30)
```

```python
# 결과 출력
# evaluate() returns the loss value & metrics values for the model
print("\n Accuracy : %.4f" %(model.evaluate(x_data,y_data, verbose=2))[1])
# 학습이 더 일어나서 과적합이 나타남
```
