```python
import numpy as np
import pandas as pd
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# seed값 설정
seed = 2020
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 데이터 입력
df_pre = pd.read_csv('../dataset/wine.csv', header=None)
df = df_pre.sample(frac=1)
dataset = df.values
X = dataset[:, 0:12]
Y = dataset[:, 12]
```

```python
# 모델 설정
model = Sequential([
    Dense(30, input_dim = 12, activation='relu'),
    Dense(12, activation='relu'),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
# 모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics =['accuracy'])
```

```python
# 모델 실행 및 저장
history = model.fit(X, Y, validation_split=0.33, epochs=2000, batch_size=500)
```

```python
# y_vloss에 데스트셋으로 실험 결과의 오차 값을 설정
y_vloss = history.history['val_loss']
```

```python
# y_acc에 학습셋으로 측정한 정확도의 값을 저장
y_acc = history.history['accuracy']
```

```python
# x 값을 지정하고 정확도를 파란색으로, 오차를 빨간색으로 표시
import matplotlib.pyplot as plt
%matplotlib inline

x_len = np.arange(len(y_acc))
plt.figure(figsize=(12,8))
plt.plot(x_len, y_vloss, "o", c="red", markersize=2)
plt.plot(x_len, y_acc, "o", c="blue", markersize=2)

plt.show()
```

```python
# y_vacc에 데이트셋으로 실험 결과의 정확도를 저장
y_vacc = history.history['val_accuracy']
```

```python
# y_loss에 학습셋으로 측정한 오차 값을 저장
y_loss=history.history['loss']
```

```python
x_len = np.arange(len(y_acc))
plt.figure(figsize=(12,8))
plt.plot(x_len, y_loss, "o", c="red", markersize=1)
plt.plot(x_len, y_vacc, "o", c="blue", markersize=1)
plt.plot(x_len, y_vloss, "o", c="orange", markersize=1)
plt.plot(x_len, y_acc, "o", c="green", markersize=1)
plt.show()
```
